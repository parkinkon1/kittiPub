## LiDAR-based Localization을 통한 GPS 데이터 보정

### 데이터 불러오기, 전처리

```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
import pykitti
import sys

%matplotlib inline
np.random.seed(777)

sys.path.append('./src')

from kalman_filters import ExtendedKalmanFilter as EKF
from utils import lla_to_enu, normalize_angles
```

```python
# Odometry Results
f = open("kitti_raw_20110930_0033.txt", 'r')
lines = f.readlines()
lines_f = [[float(i) for i in arr[:-1].split(" ")] for arr in lines]
lines_f = np.array(lines_f)

# kitti dataset
kitti_root_dir = '/datasets/kitti/raw'
kitti_date = '2011_09_30'
kitti_drive = '0033'
dataset = pykitti.raw(kitti_root_dir, kitti_date, kitti_drive)
```

```python
gt_trajectory_lla = []  # [longitude(deg), latitude(deg), altitude(meter)] x N
gt_yaws = []  # [yaw_angle(rad),] x N
gt_yaw_rates= []  # [vehicle_yaw_rate(rad/s),] x N
gt_forward_velocities = []  # [vehicle_forward_velocity(m/s),] x N

for oxts_data in dataset.oxts:
    packet = oxts_data.packet
    gt_trajectory_lla.append([
        packet.lon,
        packet.lat,
        packet.alt
    ])
    gt_yaws.append(packet.yaw)
    gt_yaw_rates.append(packet.wz)
    gt_forward_velocities.append(packet.vf)

gt_trajectory_lla = np.array(gt_trajectory_lla).T
gt_yaws = np.array(gt_yaws)
gt_yaw_rates = np.array(gt_yaw_rates)
gt_forward_velocities = np.array(gt_forward_velocities)
```

```python
# pre-processing: odometry
odom_xx = np.array([i[0] for i in lines_f])
odom_yy = np.array([i[1] for i in lines_f])
odom_x = np.array([])
odom_y = np.array([])
for idx in range(len(odom_xx)):
    odom_x = np.append(odom_x, odom_xx[idx])
    odom_y = np.append(odom_y, odom_yy[idx])
    if idx < len(odom_xx)-1:
        odom_x = np.append(odom_x, (odom_xx[idx] + odom_xx[idx+1])/2)
        odom_y = np.append(odom_y, (odom_yy[idx] + odom_yy[idx+1])/2)
        
while len(odom_x) < len(gt_yaws):
    odom_x = np.append(odom_x, odom_x[-1])
    odom_y = np.append(odom_y, odom_y[-1])

# pre-processing: gps pose
origin = gt_trajectory_lla[:, 0]  # set the initial position to the origin
gt_trajectory_xyz = lla_to_enu(gt_trajectory_lla, origin)
xs, ys, _ = gt_trajectory_xyz

# odometry-gps alignment
base = [100, len(xs)//2] # 100, 795
odom_base_x = odom_x[base]
odom_base_y = odom_y[base]
gps_base_x = xs[base]
gps_base_y = ys[base]

v1 = np.array([odom_base_x[1]-odom_base_x[0], odom_base_y[1]-odom_base_y[0]])
v2 = np.array([gps_base_x[1]-gps_base_x[0], gps_base_y[1]-gps_base_y[0]])
diff_x = odom_base_x[0] - gps_base_x[0]
diff_y = odom_base_y[0] - gps_base_y[0]
rot = np.arccos(sum(v1*v2) / (LA.norm(v1)*LA.norm(v2)))
rotMatrix = np.array([[np.cos(rot), -np.sin(rot)], [np.sin(rot), np.cos(rot)]])


odom_trans = np.array([[odom_x[idx], odom_y[idx]] for idx in range(len(odom_x))]) - np.array([odom_base_x[0], odom_base_y[0]])
odom_trans = np.matmul(odom_trans, rotMatrix) + np.array([gps_base_x[0], gps_base_y[0]])

plt.plot(odom_trans[:, 0], odom_trans[:, 1], label='odom_trans')
plt.plot(odom_x, odom_y, label='odom', alpha=0.4)
plt.plot(xs, ys, label='gps', alpha=0.4)
plt.plot(odom_base_x, odom_base_y, label='base_odom')
plt.plot(gps_base_x, gps_base_y, label='base_gps')

plt.legend()
plt.show()
```

### 성능 비교를 위한 칼만 필터 적용

#### 변수 초기화

```python
# initial state x_0
initial_yaw_std = np.pi
initial_yaw = gt_yaws[0] + np.random.normal(0, initial_yaw_std)

x = np.array([
    obs_trajectory_xyz[0, 0],
    obs_trajectory_xyz[1, 0],
    initial_yaw
])

# covariance for initial state estimation error (Sigma_0)
P = np.array([
    [xy_obs_noise_std ** 2., 0., 0.],
    [0., xy_obs_noise_std ** 2., 0.],
    [0., 0., initial_yaw_std ** 2.]
])

# Prepare measuerment error covariance Q
Q = np.array([
    [xy_obs_noise_std ** 2., 0.],
    [0., xy_obs_noise_std ** 2.]
])

# Prepare state transition noise covariance R
R = np.array([
    [forward_velocity_noise_std ** 2., 0., 0.],
    [0., forward_velocity_noise_std ** 2., 0.],
    [0., 0., yaw_rate_noise_std ** 2.]
])
```

#### 칼만 필터 적용

```python
# initialize Kalman filter
kf = EKF(x, P)

# array to store estimated 2d pose [x, y, theta]
mu_x = [x[0],]
mu_y = [x[1],]
mu_theta = [x[2],]

# array to store estimated error variance of 2d pose
var_x = [P[0, 0],]
var_y = [P[1, 1],]
var_theta = [P[2, 2],]

t_last = 0.
for t_idx in range(1, N):
    t = ts[t_idx]
    dt = t - t_last
    
    # get control input `u = [v, omega] + noise`
    u = np.array([
        obs_forward_velocities[t_idx],
        obs_yaw_rates[t_idx]
    ])
    
    # because velocity and yaw rate are multiplied with `dt` in state transition function,
    # its noise covariance must be multiplied with `dt**2.`
    R_ = R * (dt ** 2.)
    
    # propagate!
    kf.propagate(u, dt, R)
    
    # get measurement `z = [x, y] + noise`
    z = np.array([
        obs_trajectory_xyz[0, t_idx],
        obs_trajectory_xyz[1, t_idx]
    ])
    
    # update!
    kf.update(z, Q)
    
    # save estimated state to analyze later
    mu_x.append(kf.x[0])
    mu_y.append(kf.x[1])
    mu_theta.append(normalize_angles(kf.x[2]))
    
    # save estimated variance to analyze later
    var_x.append(kf.P[0, 0])
    var_y.append(kf.P[1, 1])
    var_theta.append(kf.P[2, 2])
    
    t_last = t
    

mu_x = np.array(mu_x)
mu_y = np.array(mu_y)
mu_theta = np.array(mu_theta)

var_x = np.array(var_x)
var_y = np.array(var_y)
var_theta = np.array(var_theta)
```

```python
def stds_per_interval(err, n):
    errs = np.sqrt(np.square(err))
    stds = np.array([])
    for idx in range(1,len(errs)+1):
        if idx < n//2:
            stds = np.append(stds, np.mean(errs[:idx+n//2]))
        elif idx < len(errs) - n//2:
            stds = np.append(stds, np.mean(errs[idx-n//2:idx+n//2]))
        else:
            stds = np.append(stds, np.mean(errs[idx-n//2:]))
    return stds
```

## Scenario Modeling

```python
def get_data_with_noise(title, scenario_xy_obs_noise_std=5.0, scenario_yaw_rate_noise_std=0.02, scenario_forward_velocity_noise_std=0.3):
    N = len(ts)  # number of data point
    # XY coordinate
#     scenario_xy_obs_noise_std = 5.0  # standard deviation of observation noise of x and y in meter
    scenario_xy_obs_noise = np.random.normal(0.0, scenario_xy_obs_noise_std, (2, N))  # gen gaussian noise
    scenario_obs_trajectory_xyz = gt_trajectory_xyz.copy()
    scenario_obs_trajectory_xyz[:2, :] += scenario_xy_obs_noise  # add the noise to ground-truth positions
    # yaw rate
#     scenario_yaw_rate_noise_std = 0.02 # standard deviation of yaw rate in rad/s
    scenario_yaw_rate_noise = np.random.normal(0.0, scenario_yaw_rate_noise_std, (N,))  # gen gaussian noise
    scenario_obs_yaw_rates = gt_yaw_rates.copy() + scenario_yaw_rate_noise  # add the noise to ground-truth positions
    # forward velocity
#     scenario_forward_velocity_noise_std = 0.3 # standard deviation of forward velocity in m/s
    scenario_forward_velocity_noise = np.random.normal(0.0, scenario_forward_velocity_noise_std, (N,))  # gen gaussian noise
    scenario_obs_forward_velocities = gt_forward_velocities.copy() + scenario_forward_velocity_noise  # add the noise to ground-truth positions

    # plot
    fig, ax = plt.subplots(1, 3, figsize=(18, 4))
    xs, ys, _ = gt_trajectory_xyz
    ax[0].set_title(title)
    ax[0].plot(xs, ys, lw=2, label='ground-truth trajectory')
    xs, ys, _ = scenario_obs_trajectory_xyz
    ax[0].plot(xs, ys, lw=0, marker='.', markersize=3, alpha=0.4, label='observed trajectory')
    ax[0].set_xlabel('X [m]')
    ax[0].set_ylabel('Y [m]')
    ax[0].legend()
    ax[0].grid();

    ax[1].plot(ts, gt_yaw_rates, lw=1, label='ground-truth')
    ax[1].plot(ts, scenario_obs_yaw_rates, lw=0, marker='.', alpha=0.4, label='observed')
    ax[1].set_xlabel('time elapsed [sec]')
    ax[1].set_ylabel('yaw rate [rad/s]')
    ax[1].legend();

    ax[2].plot(ts, gt_forward_velocities, lw=1, label='ground-truth')
    ax[2].plot(ts, scenario_obs_forward_velocities, lw=0, marker='.', alpha=0.4, label='observed')
    ax[2].set_xlabel('time elapsed [sec]')
    ax[2].set_ylabel('forward velocity [m/s]')
    ax[2].legend();
    
    return scenario_obs_trajectory_xyz, scenario_obs_yaw_rates, scenario_obs_forward_velocities
```

```python
def apply_kalman(obs_trajectory_xyz, obs_yaw_rates, obs_forward_velocities, 
                 xy_obs_noise_std=5.0, yaw_rate_noise_std=0.02, forward_velocity_noise_std=0.3, show_plot=True):
    scenario_obs_trajectory_xyz = obs_trajectory_xyz.copy()
    scenario_obs_yaw_rates = obs_yaw_rates.copy()
    scenario_obs_forward_velocities = obs_forward_velocities.copy()
    N = len(scenario_obs_forward_velocities)
    initial_yaw_std = np.pi
    initial_yaw = gt_yaws[0] + np.random.normal(0, initial_yaw_std)
    x = np.array([scenario_obs_trajectory_xyz[0, 0],scenario_obs_trajectory_xyz[1, 0],initial_yaw])
    P = np.array([[xy_obs_noise_std ** 2., 0., 0.],[0., xy_obs_noise_std ** 2., 0.],[0., 0., initial_yaw_std ** 2.]])
    Q = np.array([[xy_obs_noise_std ** 2., 0.],[0., xy_obs_noise_std ** 2.]])
    R = np.array([[forward_velocity_noise_std ** 2., 0., 0.],[0., forward_velocity_noise_std ** 2., 0.],[0., 0., yaw_rate_noise_std ** 2.]])

    kf = EKF(x, P)
    mu_x = [x[0],]
    mu_y = [x[1],]
    mu_theta = [x[2],]
    var_x = [P[0, 0],]
    var_y = [P[1, 1],]
    var_theta = [P[2, 2],]

    t_last = 0.
    for t_idx in range(1, N):
        t = ts[t_idx]
        dt = t - t_last
        u = np.array([scenario_obs_forward_velocities[t_idx], scenario_obs_yaw_rates[t_idx]])
        R_ = R * (dt ** 2.)
        kf.propagate(u, dt, R)

        z = np.array([scenario_obs_trajectory_xyz[0, t_idx],scenario_obs_trajectory_xyz[1, t_idx]])
        kf.update(z, Q)

        mu_x.append(kf.x[0])
        mu_y.append(kf.x[1])
        mu_theta.append(normalize_angles(kf.x[2]))

        var_x.append(kf.P[0, 0])
        var_y.append(kf.P[1, 1])
        var_theta.append(kf.P[2, 2])

        t_last = t

    mu_x = np.array(mu_x)
    mu_y = np.array(mu_y)
    mu_theta = np.array(mu_theta)
    var_x = np.array(var_x)
    var_y = np.array(var_y)
    var_theta = np.array(var_theta)
    
    # plot
    if show_plot:
        fig, ax = plt.subplots(1, 3, figsize=(18, 4))
        xs, ys, _ = gt_trajectory_xyz
        ax[0].plot(xs, ys, lw=2, label='ground-truth trajectory')
        xs, ys, _ = scenario_obs_trajectory_xyz
        ax[0].plot(xs, ys, lw=0, marker='.', markersize=2, alpha=1., label='observed trajectory')
        ax[0].plot(mu_x, mu_y, lw=2, label='estimated trajectory', color='r')
        ax[0].set_title("trajectory evaluation")
        ax[0].set_xlabel('X [m]')
        ax[0].set_ylabel('Y [m]')
        ax[0].legend()
        ax[0].grid()
        ax[1].set_title("x coordinates evaluation")
        ax[1].plot(ts, gt_trajectory_xyz[0], lw=2, label='ground-truth')
        ax[1].plot(ts, scenario_obs_trajectory_xyz[0], lw=0, marker='.', markersize=3, alpha=0.4, label='observed')
        ax[1].plot(ts, mu_x, lw=1, label='estimated', color='r')
        ax[1].set_xlabel('time elapsed [sec]')
        ax[1].set_ylabel('X [m]')
        ax[1].legend();
        ax[2].set_title("x coordinates evaluation")
        ax[2].plot(ts, mu_x - gt_trajectory_xyz[0], lw=1.5, label='estimation error')
        ax[2].plot(ts, np.sqrt(var_x), lw=1.5, label='estimated 1-sigma interval', color='darkorange')
        ax[2].plot(ts, -np.sqrt(var_x), lw=1.5, label='', color='darkorange')
        ax[2].set_xlabel('time elapsed [sec]')
        ax[2].set_ylabel('X estimation error [m]')
        ax[2].legend();
    
    return mu_x, mu_y, mu_theta, var_x, var_y, var_theta
```

```python
def averageVar(observed, gt, gap):
    err = observed - gt
    var_ = np.square(err)
    var = var_.copy()
    for idx in range(1,len(var)):
        s = idx - gap
        if s < 0: s = 0
        var[idx] = np.mean(var_[s:idx])
    return var
```

### 데이터 가공

```python
def kalmanFusion(gps, gps_var, odom, odom_var_const):
    odom_fusion = odom.copy()
    fusion = gps.copy()
    fusion_var = gps_var.copy()
    
    def update_value(mu1, var1, mu2, var2):
        mu = (mu1*var2 + mu2*var1) / (var1 + var2)
        var = (var1 * var2) / (var1 + var2)
        return mu, var
    
    for idx in range(len(fusion)):
        fusion[idx], fusion_var[idx] = update_value(fusion[idx], fusion_var[idx], odom_fusion[idx], odom_var_const)
        err = odom_fusion[idx] - fusion[idx]
        odom_fusion[idx:] -= err
        
    return fusion, fusion_var, odom_fusion
```

```python
def kalmanFusion_forTunnel(start, end, gps1, gps_var1, gps2, gps_var2, odom, odom_var_const):
    odom_fusion = odom.copy()
    fusion = odom.copy()
    fusion_var = odom.copy()
    
    def update_value(mu1, var1, mu2, var2):
        mu = (mu1*var2 + mu2*var1) / (var1 + var2)
        var = (var1 * var2) / (var1 + var2)
        return mu, var
    
    for gps_idx in range(len(gps1)):
        idx = gps_idx
        fusion[idx], fusion_var[idx] = update_value(gps1[gps_idx], gps_var1[gps_idx], odom_fusion[idx], odom_var_const)
        err = odom_fusion[idx] - fusion[idx]
        odom_fusion[idx:] -= err
        
    for idx in range(start, end):
        fusion[idx], fusion_var[idx] = odom_fusion[idx], odom_var_const
        
    for gps_idx in range(len(gps2)):
        idx = gps_idx + end
        fusion[idx], fusion_var[idx] = update_value(gps2[gps_idx], gps_var2[gps_idx], odom_fusion[idx], odom_var_const)
        err = odom_fusion[idx] - fusion[idx]
        odom_fusion[idx:] -= err
        
    return fusion, fusion_var, odom_fusion
```
